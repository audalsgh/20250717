미션1. 동영상을 업로드하여 YOLOv8에 넣어보기.<br>
결과 영상은 기본적으로 runs/detect/predict에 동영상 이름과 똑같은 .avi파일이 생긴다.

```python
!pip install ultralytics # 사진 업로드를 위한 라이브러리 설치, 셀 명령어 실행시엔 반드시 !를 붙힌다.

from google.colab import files
from ultralytics import YOLO # COCO 데이터셋에서 사전 훈련된 YOLOv8n 모델 로드

# 모델 로드후, 정보 표시 (선택사항)
model = YOLO("yolov8n.pt")
model.info()

# 동영상 업로드 후 추론 시작
print("동영상을 선택하세요")
uploaded = files.upload()
video_path = list(uploaded.keys())[0]
results = model(video_path, save=True)

# 영상 다운을 위한 라이브러리들
import os
import glob

# 📂 결과 영상 파일 경로 탐색
result_folders = glob.glob("runs/detect/predict*")
if result_folders:
    latest_folder = max(result_folders, key=os.path.getctime)
    filename = os.path.basename(video_path)
    result_video_path = os.path.join(latest_folder, filename)

    print(f"✅ 결과 영상이 저장되었습니다: {result_video_path}")

    # 다운로드 버튼
    files.download(result_video_path)
else:
    print("❌ 결과 영상이 저장되지 않았습니다.")
```

미션2. 유튜브 링크를 입력하여 똑같이 YOLOv8을 사용해보자.

```python
!pip install ultralytics # 사진 업로드를 위한 라이브러리 설치, 셀 명령어 실행시엔 반드시 !를 붙힌다.
!pip install yt-dlp  # 유튜브 영상을 다운받기위한 yt-dlp 설치

from google.colab import files  # 구글 코랩에 파일 업로드를 하기위한 라이브러리
from ultralytics import YOLO # COCO 데이터셋에서 사전 훈련된 YOLOv8n 모델 로드
import yt_dlp
import tempfile  # 유튜브 링크로 다운받고 빠른 처리를 위한 설정을 하는 용도

youtube_url = input("🎥 유튜브 영상 주소를 입력하세요: ").strip()  # https://www.youtube.com/watch?v=54nVQ-8oUzs 사용할 예정

# yt-dlp 다운로드 옵션
temp_dir = tempfile.gettempdir()
ydl_opts = {
        'format': 'mp4/best[height<=480]',  # 480p로 제한 (빠른 처리)
        'outtmpl': tempfile.gettempdir() + '/temp_video.%(ext)s',
        'quiet': True,
    }

# 유튜브 링크로 영상 가져오기
try:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        print("⬇️ 유튜브 영상 다운로드 중...")
        info = ydl.extract_info(youtube_url, download=True)
        video_path = ydl.prepare_filename(info)
        print(f"✅ 다운로드 완료: {video_path}")
except Exception as e:
    raise RuntimeError(f"❌ 다운로드 실패: {e}")

# 🔍 YOLOv8 모델 로드 및 추론
print("🔎 YOLOv8 객체 탐지 실행 중...")
model = YOLO("yolov8n.pt")
results = model(video_path, save=True, project="runs/detect", exist_ok=True)

# 영상 다운을 위한 라이브러리들
import os
import glob

# 📂 결과 영상 파일 경로 탐색
result_folders = glob.glob("runs/detect/predict*")
if result_folders:
    latest_folder = max(result_folders, key=os.path.getctime)
    filename = os.path.basename(video_path)
    result_video_path = os.path.join(latest_folder, filename)
    print(f"✅ 결과 영상이 저장되었습니다: {result_video_path}")
    print(f"📎 수동 다운로드 경로:\n{result_video_path}")  # 용량이 크면 구글 colab에서 다운이 실패할수있어서 추가함.

    # 다운로드 버튼
    files.download(result_video_path)
else:
    print("❌ 결과 영상이 저장되지 않았습니다.")
```

미션3. 첫번째 미션 코드에서 YOLOv11로 바꿔보고,<br>
결과영상 다운로드 오류를 해결해보고자 했으나, colab의 ram사용이 다해 테스트는 못해봄

```python
# 최신 YOLOv11을 위한 개발 버전 설치
!pip uninstall ultralytics -y
!pip install git+https://github.com/ultralytics/ultralytics.git

from google.colab import files
from ultralytics import YOLO # COCO 데이터셋에서 사전 훈련된 YOLOv8n 모델 로드

# 모델 로드후, 정보 표시 (선택사항)
model = YOLO("https://github.com/ultralytics/assets/releases/download/v11.0/yolov11n.pt")
model.info()

# 동영상 업로드 후 추론 시작
print("동영상을 선택하세요")
uploaded = files.upload()
video_path = list(uploaded.keys())[0]
results = model(video_path, save=True)

# 영상 다운을 위한 라이브러리들
import os
import glob

# 📂 결과 영상 파일 경로 탐색
result_folders = glob.glob("runs/detect/predict2*")
if result_folders:
    latest_folder = max(result_folders, key=os.path.getctime)

    # ✅ 자동으로 해당 폴더에서 가장 최근에 저장된 avi 파일을 찾아서 다운로드
    result_mp4s = glob.glob(os.path.join(latest_folder, '*.avi'))
    if result_mp4s:
        result_mp4s.sort(key=os.path.getmtime, reverse=True)  # 최신 파일 우선
        result_video_path = result_mp4s[0]

        print(f"✅ 실제 저장된 결과 영상: {result_video_path}")
        files.download(result_video_path)
    else:
        print("❌ 결과 폴더에 mp4 파일이 없습니다.")
else:
    print("❌ 결과 영상이 저장되지 않았습니다.")
```
